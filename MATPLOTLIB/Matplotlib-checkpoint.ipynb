```python
# AREA or STACK PLOT
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area = [2,3,2,5,4]
area1 = [2,3,4,5,6]
area2 = [1,3,2,4,2]
l = ["Area","Area1","Area2"]
c = ['r','g','y']
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Python")
plt.stackplot(x,area,area1,area2,labels=l,colors=c,baseline="zero") #winle / sym
plt.legend()
plt.show()
```

```python
# STEP PLOT
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,7,3,5,4,6]

plt.title("Python")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.step(x,y,color='r',marker='o',ms=10,mfc='b',label="Python")
plt.legend(loc=1)
plt.grid()
plt.show()
```

```python
# FILL BETWEEN PLOT
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.title("Python")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,color='r')

# plt.fill_between(x=[2,4],y1=2,y2=5,color='g')
plt.fill_between(x,y,color='g',where=(x>=2) & (x<=4),alpha=0.3)
plt.show()
```
