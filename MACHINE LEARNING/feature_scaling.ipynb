Standard Scaler (z) = (X - mean)/(standard deviation of column)

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
import pandas as pd
# scaler = StandardScaler()
# x_scaled = scaler.fit_transform()

# scaler = MinMaxScaler()
# x_scaled = scaler.fit_transform()

data = {
    'StudyHours' : [1,2,3,4,5],
    'TestScore' : [40,50,60,70,80]
}

df = pd.DataFrame(data)

std_scaler = StandardScaler()
std_scaled = std_scaler.fit_transform(df)
print('\nStandard Scaler Output')
print(pd.DataFrame(std_scaled,columns=['StudyHours','TestScore']))
```

Scaled = (X - xmin)/(Xmax-Xmin)

```python
min_max_scaler = MinMaxScaler()
minmax_scaled = min_max_scaler.fit_transform(df)

# ab ye min and max values se seekhega and unko rescale kardega between 0 and 1
print('\nMax Scaled Output')
print(pd.DataFrame(minmax_scaled,columns=['StudyHours','TestScore']))
```

```python
x = df[['StudyHours']]
y = df[['TestScore']]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# 20% data use hoga testing ke liye and 80% use hoga training ke liye
# random_state 42 ka matlab hota hai ki same result apko har time pe laana hai

print('Training Data',x_train)
print('Training Data',y_train)

print('\nTest Data',x_test)
print('\nTest Data',y_test)
```
