```python
import pandas as pd
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression
import numpy as np

data = pd.read_csv("student.csv")

x = data[['Hours']] # Double brackets = 2D input
y = data['Score'] # target column

model = LinearRegression()
model.fit(x,y)

predicted_score = model.predict(x)

mae = mean_absolute_error(y,predicted_score)
mse = mean_squared_error(y,predicted_score)
rmse = np.sqrt(mse)


print("Mean Absolute Error (MAE) : ",mae)
print("Mean Sqaured Error (MSE) : ",mse)
print("Root Mean Squared Error (RMSE) : ",rmse)

new_hour = float(input("Enter an hour: "))
new_input = pd.DataFrame([[new_hour]], columns=['Hours'])
new_pred = model.predict(new_input)

print(f"As per {new_hour} hours of studying you may score about {new_pred[0]}")
```
