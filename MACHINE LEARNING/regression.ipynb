Supervised Machine Learning cycle

Collect labelled data -> Train Model -> Make predictions -> Evaluate performance -> Refine model 

Regression -> Linear Regression 
1. predicting numbers
2. finds a pattern in old data
3. straight line 
y =  mx + c 


```python
from sklearn.linear_model import LinearRegression
x = [[1],[2],[3],[4],[5]]
y = [40,50,65,75,90]
model = LinearRegression()
model.fit(x,y)
# its not about the line its about the story
hours = float(input("How many hours did you studied? : "))

predicted_marks = model.predict([[hours]])
print(f"Based on your hours {hours} you may score around {predicted_marks}")
```

Logistic Regression is a classification model which allows you to predict a binary output using it's input features.

Like a student - Pass OR Fail

```python
from  sklearn.linear_model import LogisticRegression
x = [[1],[2],[3],[4],[5]]
y = [0,0,1,1,1] # result 0  Fail , 1 Pass

model = LogisticRegression()
model.fit(x,y)

hours = float(input("How many hours did you studied? : "))
result = model.predict([[hours]])[0]

if result == 1:
    print(f"based on your hours {hours}, you are likely to PASS")
else:
    print(f"Based on hours {hours} you've studied, you are likely to FAIL")
```

**K Nearest Neighbours**

KNN bade data ke saath boht slow kaam karta hai

```python
from sklearn.neighbors import KNeighborsClassifier
x = [[180,7],[200,7.5],[250,8],[300,8.5],[330,9],[360,9.5]]

# 0 - Apple, 1 - Orange
y = [0,0,0,1,1,1]

model = KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)
weight = float(input("Enter the weight in grams : "))
size = float(input("Enter the size in centimetes : "))

prediction  = model.predict([[weight,size]])[0] # here [0] because we want direct answer not the list
if prediction == 0:
    print(f"This is likely an apple")
else:
    print("This is likely an orange")
```

**Decision trees**

Is fever > 100
    yes > Cough?
       yes > flu
       No > Maybe
    No - No flu
    

```python
from sklearn.tree import DecisionTreeClassifier

x = [[7,2],[8,3],[9,8],[10,9]]
y = [0,0,1,1]

# 0 - Apple, 1 - Orange

model = DecisionTreeClassifier()
model.fit(x,y)

size = float(input("Enter the fruit size in centimetres : "))
shade = float(input("Enter the color shade (1-10) : "))

result = model.predict([[size,shade]])[0]

if result == 0:
    print("This is likely an Apple")
else:
    print("This is likely an Orange")
```
