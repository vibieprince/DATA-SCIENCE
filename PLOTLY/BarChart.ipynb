# **BAR CHART**

```python
import plotly.graph_objects as go
x = ['Giraffes','Lion','Monkey']
y = [20,14,23]
fig = go.Figure(go.Bar(x=x,y=y))
fig.show()
```

```python
fig = go.Figure([go.Bar(x=x,y=y,name="PLot 1"),go.Bar(x=x,y=[12,18,29],name="Plot2")])
fig.update_layout(title="Bar Charts",barmode="stack") # barmode = "group"
fig.show()
```

```python
x = ['Product A','Product B','Product C']
y = [20,14,23]

fig = go.Figure(go.Bar(x=x,y=y,hovertext=["27% Market Share","24% Market Share","29% Market Share"]))
fig.update_traces(marker_line_color='red',marker_line_width=2,opacity=0.6,text=y,textposition="inside",marker_color = ['lightblue','cyan','darkblue'],textfont=dict(size=30))
fig.show()
```

# **BAR CHART USING EXPRESS PLOTLY**

```python
import plotly.express as px
data = px.data.gapminder()
data.head()
```

```python
data_ = data.query("continent=='Europe' and year == '2007'")
data_
```

```python
fig = px.bar(data,x='country',y='pop',color='country',text='pop')
fig.update_traces(textposition='outside',texttemplate='%{text:.2s}',showlegend=False)
fig.show()
```
