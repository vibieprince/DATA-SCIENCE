# **Bubble Charts**
is connected with scatter plot
Let's see.......how?

```python
import plotly.graph_objects as go
x = [1,2,3,4]
y = [4,6,8,10]

fig = go.Figure(data=go.Scatter(x=x,y=y,text=['Product1','Product2','Product3','Product4'],mode='markers',marker=dict(size=[40,60,80,100],color=[10,12,14,16],opacity=[1,0.8,0.6,0.2],showscale=True)))
fig.show()
```
