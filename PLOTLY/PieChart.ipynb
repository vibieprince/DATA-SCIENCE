# **Pie Chart**

```python
import plotly.graph_objects as go
labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500,2500,1053,500]
fig = go.Figure(go.Pie(labels=labels,values=values))
fig.show()
```

```python
colors = ['red','orange','yellow','blue']
fig = go.Figure(go.Pie(labels=labels,values=values,textinfo='value',textfont_size=20,marker_colors=colors,marker_line_width=2,marker_line_color="aqua",hole=0.5))
fig.show()
```

```python
fig = go.Figure(go.Pie(labels=labels,values=values,pull=[0,0,0.1,0.3]))
fig.show()
```
